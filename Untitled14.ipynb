```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from datetime import datetime

print("Traffic accident analysis: road conditions, weather, and time of day")
```

```python

np.random.seed(42)

N = 1000
data = {
    "id": range(N),
    "latitude": np.random.uniform(32.0, 33.0, N),   # e.g., city area
    "longitude": np.random.uniform(75.0, 76.0, N),
    "time_of_day": np.random.choice(
        ["night", "early_morning", "morning", "afternoon", "evening"], N
    ),
    "weather": np.random.choice(
        ["sunny/clear", "rain", "fog", "snow", "other"], N, p=[0.5, 0.2, 0.15, 0.1, 0.05]
    ),
    "road_condition": np.random.choice(
        ["dry", "wet", "icy", "poor_surface"], N, p=[0.6, 0.2, 0.1, 0.1]
    ),
    "severity": np.random.choice([1, 2, 3, 4], N),  # 1 = minor, 4 = severe
}

df = pd.DataFrame(data)
print("Sample data shape:", df.shape)
print(df.head())
```

```python

print("\n--- Dataset Overview ---")
print("Unique weather conditions:", df["weather"].unique())
print("Unique road conditions:", df["road_condition"].unique())
print("Unique time of day:", df["time_of_day"].unique())
print("\nAccidents by weather:")
print(df["weather"].value_counts())
print("\nAccidents by road condition:")
print(df["road_condition"].value_counts())
print("\nAccidents by time of day:")
print(df["time_of_day"].value_counts())
```

```python
# 3. Analyze patterns: weather vs severity
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.barplot(
    data=df,
    x="weather",
    y="severity",
    estimator=np.mean,
    errorbar=None
)
plt.title("Average severity by weather")
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.barplot(
    data=df,
    x="road_condition",
    y="severity",
    estimator=np.mean,
    errorbar=None
)
plt.title("Average severity by road condition")
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.barplot(
    data=df,
    x="time_of_day",
    y="severity",
    estimator=np.mean,
    errorbar=None
)
plt.title("Average severity by time of day")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python


print("\n--- Accidents grouped by factors ---")

print("\nAccidents by weather:")
print(df.groupby("weather").size().sort_values(ascending=False))

print("\nAccidents by road condition:")
print(df.groupby("road_condition").size().sort_values(ascending=False))

print("\nAccidents by time of day:")
print(df.groupby("time_of_day").size().sort_values(ascending=False))

print("\nAccidents by weather + road condition:")
print(df.groupby(["weather", "road_condition"]).size())
```

```python
df["lat_bin"] = pd.cut(df["latitude"], bins=10, labels=False)
df["lon_bin"] = pd.cut(df["longitude"], bins=10, labels=False)

heatmap_df = df.groupby(["lat_bin", "lon_bin"]).size().reset_index(name="accident_count")

plt.figure(figsize=(8, 6))
pivot = heatmap_df.pivot(index="lat_bin", columns="lon_bin", values="accident_count")
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="Reds")
plt.title("Accident hotspot grid (binned lat/lon)")
plt.xlabel("Longitude bin")
plt.ylabel("Latitude bin")
plt.show()
```

```python

center_lat = df["latitude"].mean()
center_lon = df["longitude"].mean()

m = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=10,
    tiles="OpenStreetMap"
)

# Add heatmap layer (all accidents)
heat_data = df[["latitude", "longitude"]].values.tolist()
HeatMap(heat_data, radius=15, blur=10, max_zoom=1).add_to(m)

print("\nClick on the map to see the heatmap in Jupyter / HTML viewer.")
m  # display in Jupyter; or: m.save("accident_hotspots.html")
```

```python

print("\n--- Cross‑analysis: contributing factors ---")

print("\nWeather × road condition counts:")
wc = pd.crosstab(df["weather"], df["road_condition"])
print(wc)

plt.figure(figsize=(8, 5))
sns.heatmap(wc, annot=True, fmt="d", cmap="Blues")
plt.title("Accidents: weather vs road condition")
plt.show()

print("\nTime of day × road condition:")
toc_wc = pd.crosstab(df["time_of_day"], df["road_condition"])
print(toc_wc)

plt.figure(figsize=(8, 5))
sns.heatmap(toc_wc, annot=True, fmt="d", cmap="Oranges")
plt.title("Accidents: time of day vs road condition")
plt.show()
```
